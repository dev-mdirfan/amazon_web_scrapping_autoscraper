## Importing Libraries

```python
from autoscraper import AutoScraper
```

```python
url = "https://www.amazon.in/s?k=headphones"

wanted_list = ['₹1,199', 'M.R.P: ₹4,490 ', '(73% off)', "boAt BassHeads 100 in-Ear Wired Headphones with Mic (Black)", '201,677', '3.8 out of 5']
```

```python
scraper = AutoScraper()
result = scraper.build(url, wanted_list)
print(result)
```

```python
scraper.get_result_similar(url, grouped=True)
```

```python
scraper.set_rule_aliases({'rule_gq9i':'Price',
                         'rule_v3lg':'Offer',  
                          'rule_bv75':'PtoductTitle', 
                         'rule_1z10':'GlobalRating'})
scraper.keep_rules(['rule_gq9i', 'rule_bv75','rule_1z10'])
scraper.save('amazon_search')
```

```python
your_url = 'https://www.amazon.in/s?k=iphone'
result = scraper.get_result_similar(your_url, group_by_alias=True)
```
